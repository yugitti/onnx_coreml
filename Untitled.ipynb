```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append('../')

from cnn import net
%matplotlib inline
```

### configuration

```python
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
state = torch.load("../result/_temp.model")
data_dir = '../data'
use_cuda = True
kwargs = {'num_workers': 1, 'pin_memory': True} if use_cuda else {}
device = torch.device('cuda:{}'.format(0))
test_bach_size = 1
```

### setup data loader

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
testset = datasets.CIFAR10(
    root=data_dir, train=False, download=True, transform=transform
)
testloader = torch.utils.data.DataLoader(
    testset, batch_size=test_bach_size, shuffle=False, **kwargs
)
img_iter = iter(testloader)
```

### load model

```python
model = net.CNN().to(device)
model.load_state_dict(state)
_ = model.eval()
```

### predict

```python
img = next(img_iter)[0]
img_cuda = img.to(device)
img_np = img[0].permute(1,2,0).numpy()
torch_out = model(img_cuda)
result =torch_out[0].cpu().detach().numpy()
result = np.exp(result)
print(classes[np.argmax(result)])
img_np2 = img_np * 0.5 +0.5
plt.figure(figsize=(1,1))
plt.imshow(img_np2)
```

```python
torch.onnx.export(model,               # model being run
                  img_cuda,                         # model input (or a tuple for multiple inputs)
                  "cnn_test.onnx",   # where to save the model (can be a file or file-like object)
                  export_params=True,        # store the trained parameter weights inside the model file
                  opset_version=10,          # the ONNX version to export the model to
                  do_constant_folding=True,  # whether to execute constant folding for optimization
                  input_names = ['input'],   # the model's input names
                  output_names = ['output'], # the model's output names
                  dynamic_axes={'input' : {0 : 'batch_size'},    # variable lenght axes
                                'output' : {0 : 'batch_size'}})
```

```python
import onnxruntime

ort_session = onnxruntime.InferenceSession("cnn_test.onnx")

def to_numpy(tensor):
    return tensor.detach().cpu().numpy() if tensor.requires_grad else tensor.cpu().numpy()
```

```python
# compute ONNX Runtime output prediction
ort_inputs = {ort_session.get_inputs()[0].name: to_numpy(img_cuda)}
ort_outs = ort_session.run(None, ort_inputs)
```

```python
# compare ONNX Runtime and PyTorch results
np.testing.assert_allclose(to_numpy(torch_out), ort_outs[0], rtol=1e-03, atol=1e-05)

print("Exported model has been tested with ONNXRuntime, and the result looks good!")
```
